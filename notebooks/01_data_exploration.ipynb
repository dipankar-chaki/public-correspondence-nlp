### Load the Data

```python
# Preview the First Rows
import pandas as pd

df = pd.read_csv('../data/twcs/twcs.csv')
df.head()
```

### Exploratory Data Analysis

```python
# Length of Each Message
df['text_length'] = df['text'].apply(len)
df['text_length'].hist(bins=30)
```

```python
# Number of Tweets per Author
df['author_id'].value_counts().head(10).plot(kind='bar', title='Top 10 Active Users')
```

```python
# Tweets Over Time
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
df['created_at'].dt.date.value_counts().sort_index().plot(title='Tweets Over Time')
```
